```python
import networkx as nx
import requests
import json

ENDPOINT = 'http://localhost:8000/api/'
```

```python
def add_tag(name, description = None):
    data = {
        'name': name,
    }
    if description is not None:
        data['description'] = description

    response = requests.post(ENDPOINT + 'tags/new', json=data)
    return response.json()

add_tag('random', 'Random graph')
add_tag('gnp', 'Random G(n,p) graph')
```

```python
def to_dimacs(G):
    n = G.number_of_nodes()
    m = G.number_of_edges()
    lines = []
    lines.append(f"p ds {n} {m}")
    for u, v in G.edges():
        lines.append(f"{u + 1} {v + 1}")
    return "\n".join(lines)

for n in [10, 50, 100]:
    for avgdeg in [2, 5, 10, 15, 20, 50]:
        if avgdeg * 3 >= n:
            continue

        for _ in range(10):
            p = avgdeg / (n - 1)
            G = nx.gnp_random_graph(n, p)
            dimacs = to_dimacs(G)

            req = {
                "name": f"gnp",
                "description": f"Random G(n, p) graph with n={n} and p={p:.2E}",
                "data": dimacs,
                "submitted_by": "manpen",
                "tags": ["random", "gnp"],
            }
            
            url = ENDPOINT + 'instances/new'
            x = requests.post(url, json = req).json()
            print(x)

```
